```python
import pandas as pd

Location = "titanic3.xls"
df = pd.read_excel(Location)

df.head()
```

```python
def survivalprobability(tclass, gender, years_old, survived):
    return tclass*gender*years_old*survived
```

```python
total_passengers = len(df)

total_women = len(df[df['sex'] == 'female'])
total_men = len(df[df['sex'] == 'male'])
total_survived = len(df[df['survived'] == 1])
total_over18 = len(df[df["age"] > 18])
total_firstclass = len(df[df["pclass"] == 1])
total_2ndclass = len(df[df["pclass"] == 2])
total_3rdclass = len(df[df["pclass"] == 3])

womensurvived = len(df[(df['sex'] == 'female') & (df['survived'] == 1)])
mensurvived = len(df[(df['sex'] == 'male') & (df['survived'] == 1)])
firstclasssurvived = len(df[(df['pclass'] == 1) & (df['survived'] == 1)])
secondclasssurvived = len(df[(df['pclass'] == 2) & (df['survived'] == 1)])
thirdclasssurvived = len(df[(df['pclass'] == 3) & (df['survived'] == 1)])
adultssurvived = len(df[(df['age'] > 18) & (df['survived'] == 1)])
childsurvived = len(df[(df['age'] <= 18) & (df['survived'] == 1)])
```

```python
p_second_class = secondclasssurvived/total_survived
p_is_male = mensurvived/total_survived
p_child = childsurvived/total_survived
p_passengersurvived = total_survived/total_passengers

print (survivalprobability(p_second_class, p_is_male, p_child,p_passengersurvived))
```

```python
p_third_class = thirdclasssurvived/total_survived
p_is_female = womensurvived/total_survived
p_adult = adultssurvived/total_survived
p_passengersurvived = total_survived/total_passengers

print (survivalprobability(p_third_class, p_is_female, p_adult,p_passengersurvived))
```
